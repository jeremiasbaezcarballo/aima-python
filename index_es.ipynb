# Índice de carpeta AIMA Python

Bienvenido al repositorio de código Python de AIMA. Debería ver este cuaderno de índice si hizo clic en el botón **Iniciar carpeta** en [repository](https://github.com/aimacode/aima-python). Si está viendo este cuaderno directamente en Github, le sugerimos que utilice el botón **Iniciar Carpeta**. Binder le permite experimentar con todo el código en el navegador sin la necesidad de instalar nada en su máquina local. A continuación se muestra la lista de cuadernos que le ayudarán a navegar por los diferentes cuadernos disponibles.

Si es completamente nuevo en AIMA Python o Jupyter Notebooks, le sugerimos que comience con el Cuaderno de introducción.
